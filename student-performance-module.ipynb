```python
def random_sample(df):
    sampled_list=[]
    for x in range(10):
        temp=np.array(df.sample())
    sampled_list.append(temp)
    
    new_row=[]
    for x in sampled_list:
        for y in x:
            new_row.append(y)
    
    age=[]
    gender=[]
    hs_type=[]
    scholarship=[]
    work=[]
    activity=[]
    partner=[]
    salary=[]
    transport=[]
    living=[]
    mother_edu=[]
    father_edu=[]
    no_siblings=[]
    kids=[]
    mother_job=[]
    father_job=[]
    study_hrs=[]
    read_freq=[]
    read_freq_sci=[]
    attend_dept=[]
    impace=[]
    attend=[]
    prep_study=[]
    prep_exam=[]
    notes=[]
    listens=[]
    likes_discuss=[]
    classroom=[]
    cuml_gpa=[]
    exp_gpa=[]
    course_id=[]
    grade=[]
    
    for x in range(len(new_row)):
        age.append(new_row[x][1])
        gender.append(new_row[x][2])
        hs_type.append(new_row[x][3])
        scholarship.append(new_row[x][4])
        work.append(new_row[x][5])
        activity.append(new_row[x][6])
        partner.append(new_row[x][7])
        salary.append(new_row[x][8])
        transport.append(new_row[x][9])
        living.append(new_row[x][10])
        mother_edu.append(new_row[x][11])
        father_edu.append(new_row[x][12])
        no_siblings.append(new_row[x][13])
        kids.append(new_row[x][14])
        mother_job.append(new_row[x][15])
        father_job.append(new_row[x][16])
        study_hrs.append(new_row[x][17])
        read_freq.append(new_row[x][18])
        read_freq_sci.append(new_row[x][19])
        attend_dept.append(new_row[x][20])
        impace.append(new_row[x][21])
        attend.append(new_row[x][22])
        prep_study.append(new_row[x][23])
        prep_exam.append(new_row[x][24])
        notes.append(new_row[x][25])
        listens.append(new_row[x][26])
        likes_discuss.append(new_row[x][27])
        classroom.append(new_row[x][28])
        cuml_gpa.append(new_row[x][29])
        exp_gpa.append(new_row[x][30])
        course_id.append(new_row[x][31])
        grade.append(new_row[x][32])
    
    age=stats.mean(age)
    gender=stats.median(gender)
    hs_type=stats.mode(hs_type)
    scholarship=stats.median(scholarship)
    work=stats.mode(work)
    activity=stats.mean(activity)
    partner=stats.mode(partner)
    salary=stats.mode(salary)
    transport=stats.mean(transport)
    living=stats.median(living)
    mother_edu=stats.mean(mother_edu)
    father_edu=stats.mode(father_edu)
    no_siblings=stats.mean(no_siblings)
    kids=stats.median(kids)
    mother_job=stats.median(mother_job)
    father_job=stats.mode(father_job)
    study_hrs=stats.mean(study_hrs)
    read_freq=stats.mode(read_freq)
    read_freq_sci=stats.median(read_freq_sci)
    attend_dept=stats.median(attend_dept)
    impace=stats.mode(impace)
    attend=stats.mean(attend)
    prep_study=stats.mode(prep_study)
    prep_exam=stats.median(prep_exam)
    notes=stats.mean(notes)
    listens=stats.mean(listens)
    likes_discuss=stats.median(likes_discuss)
    classroom=stats.mode(classroom)
    cuml_gpa=stats.mean(cuml_gpa)
    exp_gpa=stats.median(exp_gpa)
    course_id=stats.mean(course_id)
    grade=stats.mean(grade)
    
    new_row=[    age,
    gender,
    hs_type,
    scholarship,
    work,
    activity,
    partner,
    salary,
    transport,
    living,
    mother_edu,
    father_edu,
    no_siblings,
    kids,
    mother_job,
    father_job,
    study_hrs,
    read_freq,
    read_freq_sci,
    attend_dept,
    impace,
    attend,
    prep_study,
    prep_exam,
    notes,
    listens,
    likes_discuss,
    classroom,
    cuml_gpa,
    exp_gpa,
    course_id,
    grade]
    
    new_row = [int(x) for x in new_row]
    return new_row
#     new_row.insert(0,len(df2)+1)
#     df2.loc[len(df2)]=new_row
```
